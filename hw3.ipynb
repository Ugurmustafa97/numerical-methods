```python
import numpy as np
from matplotlib import pyplot as plt
from math import *
```

```python
def rungeKutta(F,x,y,xStop,h):
    X = []
    Y = []
    
    X.append(x)
    Y.append(y)
    while x < xStop:
        h = min(h,xStop -x)
        K0 = h * F(x,y)
        K1 = h * F(x + h/2.0, y + K0/2.0)
        K2 = h * F(x + h/2.0, y + K1/2.0)
        K3 = h * F(x + h, y + K2)
        y = y + (K0 + 2.0*K1 + 2.0*K2 + K3)/6.0
        x = x + h
        X.append(x)
        Y.append(y)
        
    return np.array(X), np.array(Y)
```

```python
def F(x,y):
    F = np.zeros((2))
    F[0] = y[1]
    F[1] = -(9.81/0.5) * np.sin(y[0])
    return F
```

```python
x = 0.0                     # Start of integration 
xStop = 5.0                 # End of integration
y = np.array([np.pi/3, 0])  # Initial values of y
h = 0.01                    # Step size
freq = 1                    # Printout frequency
```

```python
X, Y = rungeKutta(F,x,y,xStop,h)
```

```python
plt.plot(X,Y[:,0])
plt.title(r'Time vs Angular Displacement with Initial $\theta = \pi/3$')
plt.xlabel('time(sec)')
plt.ylabel(r'$\theta(rad)$')
```

```python
plt.plot(X,Y[:,1])
plt.title(r'Time vs Angular Velocity with Initial $\theta = \pi/3$')
plt.xlabel('time(sec)')
plt.ylabel(r'$\omega(rad/s)$')
```
